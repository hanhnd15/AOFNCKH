```python
import re

def remove_special_characters(text):
    # Lo·∫°i b·ªè c√°c k√Ω t·ª± ƒë·∫∑c bi·ªát v√† bi·ªÉu t∆∞·ª£ng c·∫£m x√∫c
    cleaned_text = re.sub(r'[^\w\s]', '', text)
    return cleaned_text
```
