```python
import re

def remove_special_characters(text):
    # Lo·∫°i b·ªè c√°c k√Ω t·ª± ƒë·∫∑c bi·ªát v√† bi·ªÉu t∆∞·ª£ng c·∫£m x√∫c
    cleaned_text = re.sub(r'[^\w\s]', '', text)
    return cleaned_text
```

```python
from apify_client import ApifyClient

# Initialize the ApifyClient with your API token
client = ApifyClient(API_TOKEN)
# Use api apify/facebook-groups-scraper
# Prepare the Actor input
run_input = {
    "startUrls": [
        { "url": "https://www.facebook.com/groups/hanoiangiodau" }

    ],
    "resultsLimit": NUMBER_POST_MAX,
    # t√¨m ki·∫øm theo b√†i vi·∫øt li√™n quan
    "viewOption": "TOP_POSTS",
}

# Run the Actor and wait for it to finish
run = client.actor("2chN8UQcH1CfxLRNE").call(run_input=run_input)
post_list = []
# Fetch and print Actor results from the run's dataset (if there are any)
for item in client.dataset(run["defaultDatasetId"]).iterate_items():
    print(item)
    post_list.append(item)
```

```python
print((post_list[0]).keys())
urls_post = []
for item in post_list:
  urls_post.append(item['url'])

print(len(post_list))
```

```python
# X√¢y d·ª±ng dictionary ch·ª©a c√°c URL
start_urls = [{"url": url} for url in urls_post]
# Prepare the Actor input apify/facebook-comments-scraper
# thay "viewOption" ƒë·ªÉ thay ƒë·ªïi ch·∫ø ƒë·ªô l·ªçc comments
# "viewOption": "RANKED_THREADED" li√™n quan nh·∫•t
# "viewOption": "RECENT_ACTIVITY" m·ªõi nh·∫•t
# "viewOption": "RANKED_UNFILTERED" m·∫∑c ƒë·ªãnh c·ªßa fb
run_input = {
    "startUrls": start_urls,
    "resultsLimit": NUMBER_COMMENTS_MAX,
    "includeNestedComments": True,
    "viewOption": "RANKED_THREADED",
}

# Run the Actor and wait for it to finish
run = client.actor("us5srxAYnsrkgUv2v").call(run_input=run_input)
cmt_list = []
# Fetch and print Actor results from the run's dataset (if there are any)
for item in client.dataset(run["defaultDatasetId"]).iterate_items():
    print(item)
    cmt_list.append(item)
```
